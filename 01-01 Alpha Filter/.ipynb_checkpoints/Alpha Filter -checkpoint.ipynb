## The $\alpha$ filter

A static system or time invariant is a system that doesn't change its state over a reasonable period

### Motivational Example
Consider a rock, and the system state is the weight of the rock. The dynamic model ofthe system is constant since we assume that the weight doesn't change over short periods. 

Lets consider an instrument that is low accuracy an low precision, and with it we want to compute the rock weigh. The measures and the true value of the sensor could be something like the next image


<div>

    <p style='text-align:center;'>Figure 1. Sensor Data </p>
</div>


At time $n$ the estimate $\hat{x}_{n,n}$ would be the average of the previous measurements

\begin{equation}
\hat{x}_{n,n} = \frac{1}{n}\Big(z_1 + z_2 + \cdots + z_n\Big) = \frac{1}{n}\sum_{i=1}^{n}(z_i)\hspace{10cm}\text{Eq.1}
\end{equation}

Some important notations 
- $x$ is the true value .
- $z_n$ mesure in the time $n$.
- $\hat{x}_{n,n}$ Estimate of $x$ at time $n$ (the estimate after taking the measurement $z_n$).
- $\hat{x}_{n+1,n}$ Estimate of the future state ($n+1$) of $x$. The estimate is made at the time $n$. In other words $\hat{x}_{n+1}$ is a predicted state or extrapolated state.
- $\hat{x}_{n-1,n-1}$ Is the estimate of $x$ at time $n-1$ (the estimate is made after takin the measurment $z_{n-1}$
- $\hat{x}_{n,n-1}$ Is a prior prediction - the estimate of the state at time $n$. The predictio is made at time $n-1$

Observe that in this particuar case the dynamic mode is static or constant, since the weight of the sotone doesn't change. Hence
\begin{equation}
\hat{x}_{n+1,n} = \hat{x}_{n,n}\hspace{10cm}\text{Eq.2}
\end{equation}

Note that if we would like to implement Eq.1 into an embedded system, this would require a lot of memory, hence Eq.1 has to be represented into its recursive representation.

\begin{align*}
\hat{x}_{n,n} & = \frac{1}{n}\sum_{i=1}^n(z_i)\\
& = \frac{1}{n}\Bigg(\sum_{i=1}^{n-1}z_i+z_n\Bigg)\\
& = \frac{1}{n}\Bigg(\sum_{i=1}^{n-1}z_i\Bigg) + \frac{1}{n}z_n\\
& = \Bigg(\frac{1}{n}\Bigg)\Bigg(\frac{n-1}{n-1}\Bigg)\sum_{i=1}^{n-1}z_i + \frac{1}{n}z_n\\
& = \Bigg(\frac{n-1}{n}\Bigg)\Bigg[\frac{1}{n-1}\sum_{i=1}^{n-1}z_i \Bigg] + \frac{1}{n}z_n\hspace{5cm}\text{Eq.3}
\end{align*}

Observe that the term within the squared brackets is equivalent to Eq.1 but in a previous time

\begin{equation}
\hat{x}_{n-1,n-1} = \frac{1}{n-1}\sum_{i=1}^{n-1}z_i \hspace{5cm} \text{Eq.4}
\end{equation}

Hence the Eq.3 can be rewritten as 

\begin{align*}
\hat{x}_{n,n} & = \frac{n-1}{n}\hat{x}_{n-1,n-1} + \frac{1}{n}z_n \\
& = \hat{x}_{n-1,n-1} -\frac{1}{n}\hat{x}_{n-1,n-1} + \frac{1}{n}z_n \hspace{5cm}\\
& = \hat{x}_{n-1,n-1} -\frac{1}{n}\hat{x}_{n-1,n-1} + \frac{1}{n}z_n \hspace{5cm} \text{Eq.5}\\
\end{align*}




Finally Eq.5 can be rewritten as

\begin{align*}
\hat{x}_{n,n} & = \hat{x}_{n-1,n-1} + \frac{1}{n}\Big(z_n - \hat{x}_{n-1,n-1}\Big)\hspace{5cm} \text{Eq.6}\\
\end{align*}

The Eq.6 is known as **STATE UPDATE EQUATION** and the term within the parenthesis is known as **MEASUREMENT RESIDUAL**. Observe that after some interations the term $1/n\rightarrow 0$, hece at some parint the contribution of new measurements will become negligible. 

For sake of simplicity let's define 

\begin{equation}
\alpha_n = \frac{1}{n} \hspace{5cm} Eq.7
\end{equation}

Hence Eq.6 can be rewritten as 
\begin{align*}
\hat{x}_{n,n} & = \hat{x}_{n-1,n-1} + \alpha_n\Big(z_n - \hat{x}_{n-1,n-1}\Big)\hspace{5cm} \text{Eq.6}\\
\end{align*}

**NOTE THAT THIS DOCUMENT EXPLAINS THE MEAN FILTER OF THIS TUTORIAL BUT IT IS EXPRESSED IN A VERY IMPORTANT EXPRESSION, BECAUSE THE STATE UPDATE EQUATION IS USED IN THE KALMAN FILTER**

## Numerical Example - 10 Iterations
### Iteration 1 and Initialization
###### Initialization 
$\hat{x}_{0,0} = 1000$
###### Precondition
$\hat{x}_{1,0} = 1000$

###### Step 1 
$z_1 = 996$
###### Step 2
Compute the gain

$\alpha_1 = \frac{1}{n} = \frac{1}{1} = 1$

Estimate the current state using the State Update Equation

$\hat{x}_{n,n}=\hat{x}_{n,n-1} + \alpha_n(z_n-\hat{x}_{n,n-1})$

$\hat{x}_{1,1}=\hat{x}_{1,0} + \alpha_1(z_1-\hat{x}_{1,0})$

$\hat{x}_{1,1}=1000 + 1(996-1000)=996$

###### Step 3

Calculate the predicted state for the next iteration using System's Dynamic MOdel. Since the system is static then 

$x_{n+1,n} = \hat{x}_{n,n}$

$x_{2,1} = \hat{x}_{1,1} = 996$

### Iteration 2 
Prior estimate

$x_{2,1} = \hat{x}_{1,1} = 996$

###### Step 1 

$z_2 = 994$

###### Step 2

Compute the gain

$\alpha_1 = \frac{1}{n} = \frac{1}{2} $

Estimate the current state using the State Update Equation

$\hat{x}_{n,n}=\hat{x}_{n,n-1} + \alpha_n(z_n-\hat{x}_{n,n-1})$

$\hat{x}_{2,2}=\hat{x}_{2,1} + \alpha_2(z_2-\hat{x}_{2,1})$

$\hat{x}_{2,2}=998 + \frac{1}{2}(994-996)=995$

###### Step 3

Calculate the predicted state for the next iteration using System's Dynamic MOdel. Since the system is static then 

$x_{n+1,n} = \hat{x}_{n,n}$

$x_{3,2} = \hat{x}_{2,2} = 995$


### Iteration 3
Prior estimate

$x_{3,2} = \hat{x}_{3,3} = 995$

###### Step 1 

$z_3 = 1021$

###### Step 2

Compute the gain

$\alpha_3 = \frac{1}{n} = \frac{1}{3} $

Estimate the current state using the State Update Equation

$\hat{x}_{n,n}=\hat{x}_{n,n-1} + \alpha_n(z_n-\hat{x}_{n,n-1})$

$\hat{x}_{3,3}=\hat{x}_{3,2} + \alpha_3(z_3-\hat{x}_{3,2})$

$\hat{x}_{3,3}=995 + \frac{1}{3}(1021-995)=1003.67$

###### Step 3

Calculate the predicted state for the next iteration using System's Dynamic MOdel. Since the system is static then 

$x_{n+1,n} = \hat{x}_{n,n}$

$x_{4,3} = \hat{x}_{3,3} = 1003.67$


### Iteration 4

$z_4 = 1000$

$\alpha_4 = \frac{1}{4}$

$\hat{x}_{4,4}=1003.67 + \frac{1}{4}(1000-1003.67)=1002.75$

$x_{5,4} = \hat{x}_{4,4} = 1002.75$

### Iteration 5

### Iteration 6

$x_{7,6} = \hat{x}_{6,6} = 1003.83$

### Iteration 7

$\hat{x}_{7,7}=1003.83 + \frac{1}{6}(983-1003.83)=1000.86$

$x_{8,7} = \hat{x}_{7,7} = 1000.86$

### Iteration 8

### Implementation 


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
z = [996, 994, 1021,1000,1002,1010,983,971,993,1023]
data = pd.DataFrame({"Samples":np.arange(1,11)})
data['z'] = np.array(z)
data["Samples"] = np.arange(1,11)
data['x_hat'] = np.zeros((10,1))
data['measurement_residual'] = np.zeros((10,1))
data['alpha']= np.zeros((10,1))
data['true_value']=1000*np.ones((10,1))
data.head(5)
```

```python
for index in range(data.shape[0]):
    if index == 0:
        previous_estimate = 1000
    else:
        previous_estimate = data.loc[index-1,"x_hat"]
    
    # alpha_n saved in the Data Frame an in variable alpha_n
    data.loc[index,"alpha"] = 1/(index+1)
    alpha_n = data.loc[index,"alpha"]
    
    # measurement residual
    meas_res = data.loc[index,"z"] - previous_estimate
    data.loc[index,"measurement_residual"] = meas_res
    
    current_state_estimate = previous_estimate + alpha_n*meas_res
    data.loc[index,"x_hat"] = current_state_estimate
    
data.head(10)
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(data["Samples"],data["z"],label='raw signal')
axes.plot(data["Samples"],data["x_hat"],color='r', label="filtered signal")
axes.plot(data["Samples"],data["true_value"],color='g', label="filtered signal")

```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(data["Samples"],data["measurement_residual"],label='raw signal')
axes.plot(data["Samples"],data["alpha"],color='r', label="filtered signal")
```

```python

dt = 0.2; 
t = np.arange(0,10+dt,dt)
#print(t)

time_span = {"time":t}

measures = np.arange(0,len(t))

data2 = pd.DataFrame({"sample":measures})
data2["time"]= t
raw_signal = [14.4 + 2*np.random.normal(0,2,1)[0] for value in data2["time"]]
data2["raw_signal"] = raw_signal
data2['filtered_signal_fnc'] = np.zeros((data2.shape[0],1))
data2['filtered_signal_own'] = np.zeros((data2.shape[0],1))
data2['x_hat_alpha'] = np.zeros((data2.shape[0],1))
data2.head(10)
```

```python

data2["time"]= t
raw_signal = [14.4 + 2*np.random.normal(0,2,1)[0] for value in data2["time"]]
data2["raw_signal"] = raw_signal
data2['filtered_signal_fnc'] = np.zeros((data2.shape[0],1))
data2['filtered_signal_own'] = np.zeros((data2.shape[0],1))
data2.head(10)
```

```python
def full_avg_filt(aIndex, aPrevAvg = 0, aCurrentSignal = 0):
    if aIndex == 0:
        lPrevAvg = 0
        lk = 1
    else: 
        lPrevAvg = aPrevAvg
        lk = aIndex + 1
    
    lAlpha = (lk -1) / lk
    return lAlpha * lPrevAvg + (1 - lAlpha)*aCurrentSignal
```

```python
for index in range(data.shape[0]):
    if index == 0: 
        prevAvg = 0
        k = 1
    else:
        prevAvg = data2.loc[index-1,"filtered_signal_own"]
        k = index + 1

    alpha = (k-1)/k
    data2.loc[index,"filtered_signal_own"] = alpha*prevAvg + (1-alpha)*data2.loc[index,"raw_signal"]
    data2.loc[index,"filtered_signal_fnc"] = full_avg_filt(aIndex = index, 
                                                       aPrevAvg = prevAvg,
                                                       aCurrentSignal = data2.loc[index,"raw_signal"])
    
```

```python
for index in range(data2.shape[0]):
    if index == 0:
        previous_estimate = 1000
    else:
        previous_estimate = data2.loc[index-1,"x_hat"]
    
    # alpha_n
    alpha_n = 1/(index+1)
    
    # measurement residual
    meas_res = data2.loc[index,"raw_signal"] - previous_estimate
    
    current_state_estimate = previous_estimate + alpha_n*meas_res
    data2.loc[index,"x_hat"] = current_state_estimate
```
